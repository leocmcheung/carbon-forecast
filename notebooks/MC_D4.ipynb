```python
import pandas as pd
```

```python
all_features = pd.read_excel('../../mcussac/data_forecast_carbon/All_features_v2.xlsx')
all_features.head()
```

```python
sector_rev = pd.read_csv('../../mcussac/data_forecast_carbon/Sectorrev.csv')
sector_rev.head()
```

```python
# drop columns
sector_rev.drop(columns=['Company', 'ISIN', 'GICS Sector Code',
       'GICS Sector Name', 'GICS Industry Group Code',
       'GICS Industry Group Name', 'GICS Industry Code', 'GICS Industry Name',
       'GICS Sub Industry Code', 'GICS Sub Industry Name', 'GICS Description',
       'Country', 'Trucost Sector',
       'Trucost Sector Revenue (USD mn)', 'FigureRestate',
       'EffectiveDate'], inplace=True)
```

```python
sector_rev.head()
```

```python
sector_rev_2020 = sector_rev[sector_rev['Financial Year']==2020]
```

```python
sector_rev_2020_formatted = pd.pivot_table(sector_rev_2020, values='Trucost Sector Revenue Percentage (%)' , columns='Trucost Sector Name', index='TCUID')
```

```python
sector_rev_2020_formatted
```

```python
all_features_with_sectorrev = pd.merge(all_features, sector_rev_2020_formatted, on='TCUID')
```

```python
all_features_with_sectorrev.shape
```

```python
all_features_with_sectorrev.to_csv('../../mcussac/data_forecast_carbon/all_features_with_sectorrev.csv')
```
