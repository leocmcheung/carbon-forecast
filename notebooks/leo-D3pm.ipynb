```python
from carbon_forecast.utils.data_related import show_sector
import pandas as pd
import numpy as np
```

```python
df = show_sector()
```

```python
df
```

```python
c_abs = df.pivot_table(index="Company", columns="Financial Year", values="C-absolute-1+2").reset_index()
c_abs["Data Type"] = "C-absolute-1+2"
cols =['Company','Data Type', 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
c_abs = c_abs[cols]
c_abs
```

```python
c_w_disclosure = df.pivot_table(index="Company", columns="Financial Year", values="Carbon-Weighted Disclosure (%)").reset_index()
c_w_disclosure["Data Type"] = "Carbon-Weighted Disclosure (%)"
c_w_disclosure = c_w_disclosure[cols]
c_w_disclosure
```

```python
revenue = df.pivot_table(index="Company", columns="Financial Year", values="Revenue (USD mn)").reset_index()
revenue["Data Type"] = "Revenue (USD mn)"
revenue = revenue[cols]
revenue
```

```python
c_abs_data = c_abs.drop(columns=["Company", "Data Type"])
c_w_disclosure_data = c_w_disclosure.drop(columns=["Company", "Data Type"])
revenue_data = revenue.drop(columns=["Company", "Data Type"])

c_abs_data = c_abs_data.iloc[:, 5:].dropna()
c_w_disclosure_data = c_w_disclosure_data.iloc[:, 5:].dropna()
revenue_data = revenue_data.iloc[:, 5:].dropna()
revenue_data
```

```python
c_abs_data_train, c_abs_data_test = c_abs_data[:120], c_abs_data[120:]
c_w_disclosure_data_train, c_w_disclosure_data_test = c_w_disclosure_data[:120], c_w_disclosure_data[120:]
revenue_data_train, revenue_data_test =revenue_data[:120], revenue_data[120:]
c_w_disclosure_data_test.shape
```

```python
X_c_abs_train, y_c_abs_train = c_abs_data_train.drop(columns=[2019,2020]), c_abs_data_train[[2019,2020]]
X_c_w_disclosure_train, y_c_w_disclosure_train = c_w_disclosure_data_train.drop(columns=[2019,2020]), c_w_disclosure_data_train[[2019,2020]]
X_revenue_train, y_revenue_train = revenue_data_train.drop(columns=[2019,2020]), revenue_data_train[[2019,2020]]
X_revenue_train.shape
```

```python
y_c_abs_train
```

```python
X_big = [X_c_abs_train, X_c_w_disclosure_train, X_revenue_train]
y_big = [y_c_abs_train]
```

```python
X_big_array = np.array(X_big).astype(np.float32)
y_big_array = np.array(y_big).astype(np.float32)
X_big_array.shape
```

```python
X_big_array_train = X_big_array.reshape(120,9,3)
y_big_array_train = y_big_array.reshape(120,2,1)
y_big_array_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Masking, Normalization
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from tensorflow.keras.losses import mean_absolute_percentage_error
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
def scheduler(epoch, lr):
    if epoch < 200:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

normalizer = Normalization(input_shape=(9,3))
normalizer.adapt(X_big_array_train)


model = Sequential([
    normalizer,
    LSTM(1000,return_sequences=True),
    LSTM(500),
    Dense(200),
    Dense(50),
    Dense(1, activation="linear")
])

es = EarlyStopping(patience=50, restore_best_weights=True)
lr = LearningRateScheduler(scheduler)
model.summary()
```

```python
model.compile(Adam(learning_rate=0.05), loss=mean_absolute_percentage_error)
```

```python
history = model.fit(x=X_big_array_train, y=y_big_array_train,epochs=10_000, callbacks=[lr, es], validation_split=0.2)
```
